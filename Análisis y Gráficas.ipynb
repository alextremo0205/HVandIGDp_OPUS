# Cargar las librerías

```python
import pandas as pd
import numpy as np
import os

import plotly.graph_objects as go
import scipy.stats
import plotly.express as px
```

```python
#Computes the 95% confidence interval for any set of data. 
#Assumes normal distribution, unknown variance, iid
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

# Programa principal

```python
Cod = {
    'Hanoi':'HAN',
    'Fossolo':'FOS',
    'Balerma':'BAL',
    'Modena':'MOD'
}
retro={
    'Hanoi':5,
    'Fossolo':20,
    'Balerma':50,
    'Modena':10
}
rangosx = {
    'HV':{
            'Hanoi':[450,500],
            'Fossolo':[450,500],
            'Balerma':[1500,2000],
            'Modena':[1500, 2000]
    },
    'IGDp':{
            'Hanoi':[0,60],
            'Fossolo':[0, 60],
            'Balerma':[0, 500],
            'Modena':[0, 500]
    }
}
rangosy = {
    'HV':{
            'Hanoi':[0.69,0.71], #Ajustado
            'Fossolo':[0.91, 0.98], #Ajustado
            'Balerma':[0.8, 1],
            'Modena':[0.8, 1]
    },
    'IGDp':{
            'Hanoi':[0,0.6], #Ajustado
            'Fossolo':[0, 0.065], #Ajustado
            'Balerma':[0, 0.065],
            'Modena':[0, 0.065]
    }
}
colors={
    'NSGAII':'firebrick',
    'OPUS':'darkblue',
    'OPT': 'green'
}
nombres = {
    'HV': 'HV',
    'IGDp':"IGD+",
    'OPUS':'NSGAII + OPUS',
    "NSGAII":"NSGAII"
}
# print(len(actual))
# print(actual.iloc[500])
```

```python
#Cargar los datos
methods = ['NSGAII', 'OPUS']
sims = list(range(1,31))

Mat_X = pd.DataFrame(data=[], columns=['Gen', metrica, 'Method', 'Sim'])

for meth in methods:
    for sim in sims:    
        actual = pd.read_csv('./'+red+" "+meth+'/'+ metrica+'_'+Cod[red]+'_'+meth+'_'+str(sim)+'.csv', header =0 ,names=['Gen', metrica])
        actual.head()
        actual['Sim'] = sim
        actual['Method'] = meth

        if metrica == 'HV':
            metr_opt = pd.DataFrame(data = [[0,actual.iloc[len(actual)-1][1], meth,sim]], columns= ['Gen', metrica, 'Method', 'Sim'])
            metr_opt = metr_opt.append(actual.iloc[len(actual)-1], ignore_index = True)
            actual = actual.iloc[0:-1]
        else:
            metr_opt = pd.DataFrame(data = [[0,0, meth,sim], [len(actual), 0, meth,sim]], columns= ['Gen', metrica, 'Method', 'Sim'])
        Mat_X = Mat_X.append(actual, ignore_index = True)
        print(metr_opt)
        #actual.tail()
```

```python
generations = list(actual['Gen'])
print(generations)
```

```python


Mat_Y = pd.DataFrame(data = [], columns=['Gen','Mean '+metrica, 'Upper', 'Lower','Method'])

for meth in methods:
    for gen in generations:
        prueba = Mat_X[(Mat_X['Gen']==gen) & (Mat_X['Method']==meth)]
        mean, plus, minus = mean_confidence_interval(prueba[metrica], 0.95)
        df_act = pd.DataFrame(data = [[gen, mean, plus, minus, meth]], columns=['Gen','Mean '+metrica, 'Lower', 'Upper','Method'])
        Mat_Y = Mat_Y.append(df_act, ignore_index = True, sort=False)
        
Mat_Y.head()
```
